```python
# Takes in a list of patients IDs .
# Returns a list of file paths fro HR 
def get_file_paths(patient_ids):
    hr_filepaths = []
    acc_pids = []
    acc_filepaths = []
    for patientid in patient_ids: # Iterate over all patient IDs.
        patient_dir = "/Users/shehjarsadhu/Desktop/UniversityOfRhodeIsland/Graduate/ResearchWBL/My_Thesis/care-portal-thesis-sub/PatientID_" + str(patientid)  
        files = os.listdir(patient_dir)
        for file in files:
            print(file)
            if file.startswith("Date_ID_HR.csv"): # Check if HR data.We dont want acceleration for now.
#                 print(file)
                path = patient_dir + "/" + file
                df = pd.read_csv(path)
                print(df["Patient.ID"].unique())
                print("path: = \n",path.split("/My_Thesis",1))
                hr_filepaths.append(path.split("/My_Thesis",1)[1])
            if file.startswith("Date_ID_Acc.csv"):
                acc_path = patient_dir + "/" + file
                df = pd.read_csv(acc_path)
                print(df["Patient.ID"].unique())
                acc_pids.append(df["Patient.ID"].unique()[0])
                print("Acc path: = \n",acc_path)
                acc_filepaths.append(acc_path.split("/My_Thesis",1)[1])
    return hr_filepaths,acc_filepaths,acc_pids
```
